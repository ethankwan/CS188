```python
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import perceptron
import pandas as pd

data1 = pd.read_csv('data1.csv')
data1.columns = ['x1', 'x2', 'y']
colormap = np.array(['r', 'b'])
data1.head()
data1.plot.scatter(0, 1)
```

```python



net = perceptron.Perceptron(max_iter=1000, verbose=0, random_state=None, fit_intercept=True, shuffle=False)

# Train the perceptron object (net)
net.fit(data1[['x1','x2']],data1['y'])
# Output the values
print ("Coefficient 0: " + str(net.coef_[0, 0]))
print ("Coefficient 1: " + str(net.coef_[0, 1]))
print ("Bias: " + str(net.intercept_))

plt.scatter(data1.x1,data1.x2, c=colormap[data1.y],s=20)

# Calc the hyperplane (decision boundary)
xmin, xmax = plt.ylim()
w = net.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(xmin, xmax)
yy = a * xx - (net.intercept_[0]) / w[1]

# Plot the hyperplane
plt.plot(xx, yy, 'k-')
plt.show()
```
